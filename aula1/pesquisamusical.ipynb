```python
import pandas as pd 
pesquisa_simulada = pd.read_csv("dataset/pesquisasimulada.csv")
pesquisa_simulada
```

```python
pesquisa_simulada.columns = ['data_hora','rock','hip_hop','mpb','reggae','funk','sertanejo','gospel','eletronica']
# removendo a coluna data hora para realizar o calculo
notas = pesquisa_simulada.drop(columns=['data_hora'])
notas
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Criar um DataFrame com mais categorias
dados_ficticios = pd.DataFrame({
    'Categoria': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'Valor': [10, 20, 15, 25, 30, 18, 22, 27, 35]
})

# Criar o gráfico de barras com mais categorias
plt.figure(figsize=(12, 6))  # Ajustar o tamanho da figura para acomodar mais barras
sns.barplot(x='Categoria', y='Valor', data=dados_ficticios)

# Adicionar título ao gráfico
plt.title('Gráfico de Barras com Mais Categorias')
plt.xlabel('Categoria')
plt.ylabel('Valor')

# Mostrar o gráfico
plt.show()
```

```python
# Cria um histograma para cada estilo musical
plt.figure(figsize=(14, 12))

for i, column in enumerate(notas.columns, 1):
    plt.subplot(3, 3, i)  # Ajusta o layout para vários subgráficos
    sns.histplot(notas[column], bins=5, kde=True, color='skyblue')
    plt.title(f'Estilo: {column}')
    plt.xlabel('Nota')
    plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
