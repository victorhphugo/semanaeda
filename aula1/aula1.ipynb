```python
import pandas as pd 
```

```python
pesquisa = pd.read_csv("dataset/pesquisaaulaestatistica.csv")
pesquisa
```

```python
pesquisa.columns = ['datahora','email','pontuacao','compreensao','topicodesafiador','sentimento']
pesquisa
```

```python
import seaborn as sns
porcetagem = pesquisa['sentimento'].value_counts(normalize=True).reset_index()
porcetagem.columns=['Sentimento','Porcentagem Total']
criar_grafico = sns.barplot(x='Sentimento', y='Porcentagem Total',data=porcetagem)
criar_grafico.bar_label(criar_grafico.containers[0],fmt='%.2f%%')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supondo que 'pesquisa' é seu DataFrame e 'sentimento' é a coluna de interesse
porcetagem = pesquisa['sentimento'].value_counts(normalize=True).reset_index()
porcetagem.columns = ['Sentimento', 'Porcentagem Total']

# Define uma lista de cores para cada barra
cores = ['#006400', '#66B2FF', '#FF0000', '#FFCC99', '#FFD700']

# Cria o gráfico com a paleta de cores
criar_grafico = sns.barplot(x='Sentimento', y='Porcentagem Total', data=porcetagem, palette=cores)

# Adiciona rótulos nas barras
criar_grafico.bar_label(criar_grafico.containers[0], fmt='%.2f%%')

# Ajusta o rótulo do eixo x
criar_grafico.set_xlabel('')

# Exibe o gráfico
plt.show()
```
